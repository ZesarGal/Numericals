## La matriz jacobiana y derivadas parciales.

La *matriz jacobiana* de una función $f:\mathbb{R}^{n}\to \mathbb{R}^{m}$ en un punto $x_{0}\in \mathbb{R}^{n}$ es la matriz $J=(J_{ij})\in \text{Mat}_{m\times n}(\mathbb{R})$ tal que
$$J_{ij}=\frac{\partial f_{i}}{\partial x_{j}}(x_0).$$

```python
import numpy as np
import scipy.optimize as opt  # import root-finding algorithm
import sympy as sp  # Python toolbox for symbolic maths
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Toolbox for rendring 3D figures
from mpl_toolkits import mplot3d  # Toolbox for rendring 3D figures
```

En el módulo **SymPy** existe el método `.Matrix()` para crear matrices simbólicas y el método `.jacobian()` para evaluar la jacobiana.

**Ejemplo.** Definamos la función $$f(r,\theta)=(r\cos(\theta), r\sin(\theta)).$$

Calculemos a la matriz jacobiana de $f$.

```python
r, t = sp.symbols('r theta')
f = sp.Matrix([r*sp.cos(t), r*sp.sin(t)])
print(f) 
```

```python
J = f.jacobian([r,t])
print(J) 
```

Tomemos el punto $(1,\pi/2)\in \mathbb{R}\times [0,2\pi)$ y evaluemos la jacobiana en tal punto.

```python
J.subs({r:1, t:sp.pi/2})
```

```python
# El determinante de la matriz jacobiana
sp.simplify(J.det())
```

### Gráficas de las derivadas

Como ejemplo, tomemos la función $f(x,y)=x^2+2xy+y^2$.

```python
f2 = 'x**2 + 2*x*y + y**2'

f, x, y = sp.symbols('f, x, y')

f2d = eval(f2)
f2d_dx = sp.diff(f2d,x)
f2d_dy = sp.diff(f2d,y)

print('La derivada parcial de ', f2d, 'con respecto a x es', f2d_dx)
print('La derivada parcial de ', f2d, 'con respecto a y es', f2d_dy)
```

```python
f, x, y = sp.symbols('f, x, y')
f2d = eval(f2) 
p1 = sp.plotting.plot3d(f2d, (x, -5, 5), (y, -5, 5),show=True,xlabel='x', 
                        ylabel='y', zlabel='f(x,y)',title='Función')
```

```python
f2d_dx = sp.diff(f2d,x)
p2 = sp.plotting.plot3d(f2d_dx, (x, -5, 5), (y, -5, 5),show=True,xlabel='x', 
                        ylabel='y', zlabel='df(x,y)/dx',title='Derivada con respecto a x')
```

```python
f2d_dy = sp.diff(f2d,y)
p3 = sp.plotting.plot3d(f2d_dy, (x, -5, 5), (y, -5, 5),show=True,xlabel='x', ylabel='y', 
                        zlabel='df(x,y)/dy',title='Derivada con respecto a y')
```
